PCA of CLIP Embedding

```python
import matplotlib.pyplot as plt
import torch

PATH = "multimodal_embeddings.pt"

data = torch.load(PATH, map_location="cpu")
embeddings = data["embeddings"].float()

# PCA to 2D
centered = embeddings - embeddings.mean(dim=0, keepdim=True)
_, _, v = torch.pca_lowrank(centered, q=2)
points_2d = centered @ v[:, :2]

plt.figure(figsize=(8, 6))
plt.scatter(points_2d[:, 0].numpy(), points_2d[:, 1].numpy(), s=10)
plt.title("PCA of multimodal embeddings")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()
```

```python
# PCA to 3D
_, _, v3 = torch.pca_lowrank(centered, q=3)
points_3d = centered @ v3[:, :3]

fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(
    points_3d[:, 0].numpy(),
    points_3d[:, 1].numpy(),
    points_3d[:, 2].numpy(),
    s=10,
)
ax.set_title("PCA of multimodal embeddings (3D)")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from IPython.display import HTML, display

rows = data["rows"]
object_ids = [str(row["object_id"]) for row in rows]
image_files = [row["filename"] for row in rows]

coords = points_3d.detach().cpu().numpy()
customdata = np.column_stack([object_ids, image_files])

fig = go.Figure(
    data=[
        go.Scatter3d(
            x=coords[:, 0],
            y=coords[:, 1],
            z=coords[:, 2],
            mode="markers",
            marker={"size": 4},
            customdata=customdata,
            hovertemplate=(
                "object_id: %{customdata[0]}<br>"
                "image_file: %{customdata[1]}<extra></extra>"
            ),
        )
    ]
)

fig.update_layout(
    title="Interactive 3D PCA",
    scene={"xaxis_title": "PC1", "yaxis_title": "PC2", "zaxis_title": "PC3"},
)

display(HTML(fig.to_html(include_plotlyjs="cdn", full_html=False)))
```
