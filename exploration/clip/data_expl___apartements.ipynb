Explore Scraped Data

```python
import json
from pathlib import Path

FILE_PATH = Path("apartements.jsonl")
RECORD_INDEX = 0

records = []
with FILE_PATH.open("r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line:
            records.append(json.loads(line))

print(json.dumps(records[RECORD_INDEX], indent=2, ensure_ascii=False))
```

Num of records

```python
count = 0
with FILE_PATH.open("r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            count += 1
print(f"Total records: {count}")
```

NULL or empty strings

```python
summary = {}

for rec in records:
    for field, value in rec.items():
        if field not in summary:
            summary[field] = {"missing_or_null": 0, "empty": 0}

        if value is None:
            summary[field]["missing_or_null"] += 1
        elif isinstance(value, str) and value.strip() == "":
            summary[field]["empty"] += 1

        
summary = {
    field: stats
    for field, stats in summary.items()
    if stats["missing_or_null"] > 0 or stats["empty"] > 0
}

if not summary:
    print("No attributes with null or empty strings found.")
else:
    for field, stats in summary.items():
        print(f"{field}: null_or_missing={stats['missing_or_null']}, empty_strings={stats['empty']}")
```

Length of Attributes

```python
all_fields = sorted({key for rec in records for key in rec.keys()})

max_length_by_field = {}
max_value_by_field = {}

for field_name in all_fields:
    max_len = 0
    max_value = None

    for rec in records:
        value = rec.get(field_name)
        value_len = 0 if value is None else len(str(value))

        if value_len > max_len:
            max_len = value_len
            max_value = value

    max_length_by_field[field_name] = max_len
    max_value_by_field[field_name] = max_value

for field_name in all_fields:
    print(f"{field_name}: max_length={max_length_by_field[field_name]}")
```
